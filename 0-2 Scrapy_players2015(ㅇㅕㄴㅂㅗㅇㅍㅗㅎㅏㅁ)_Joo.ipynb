```python
# sudkorea 3589 54/ china 5598 54

driver = webdriver.Chrome('./chromedriver')

nation = ['sudkorea', 'china']
nationid = ['3589', '5598']

nameList = []
nationList = []
ageList = []
heightList = []
capList = []
goalList = []
salaryList = []

for i in range(len(nation)):
    print(nation[i], nationid[i])
    url_base = 'https://www.transfermarkt.co.uk/'
    url_sub = nation[i] + '/kader/verein/' + nationid[i] + '/saison_id/2014/plus/1'
    driver.get(url_base + url_sub)
    html = driver.page_source
    soup = BeautifulSoup(html,'html.parser')
    for j in range(0,54):
        nationList.append(nation[i])
    for i in range(0,541):
        if i % 10 == 1:
            name = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('-',' ').split()
            nameList.append(name)
        if i % 10 == 2:
            age = soup.select('#yw1 > table > tbody > tr > td')[i].text.split()[3].replace('(','').replace(')','')
            ageList.append(age)    
        if i % 10 == 4:
            height = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace(',','').replace('m','')
            heightList.append(height)
        if i % 10 == 6:
            cap = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('-','0')
            capList.append(cap)
        if i % 10 == 7:
            goal = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('-','0')
            goalList.append(goal)
        if i % 10 == 9:
            salary = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('.','').replace('£','').replace('k','000').replace('m','000000').replace('-','0')
            salaryList.append(salary)
```

```python
# nordkorea 15457 41
nameList = []
ageList = []
heightList = []
capList = []
goalList = []
salaryList = []

html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
for i in range(0,411):
    if i % 10 == 1:
        name = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('-',' ').split()
        nameList.append(name)
    if i % 10 == 2:
        age = soup.select('#yw1 > table > tbody > tr > td')[i].text.split()[3].replace('(','').replace(')','')
        ageList.append(age)    
    if i % 10 == 4:
        height = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace(',','').replace('m','').replace(' ','-')
        heightList.append(height)
    if i % 10 == 6:
        cap = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('-','0')
        capList.append(cap)
    if i % 10 == 7:
        goal = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('-','0')
        goalList.append(goal)
    if i % 10 == 9:
        salary = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('.','').replace('£','').replace('k','000').replace('m','000000').replace('-','0')
        salaryList.append(salary)
```
