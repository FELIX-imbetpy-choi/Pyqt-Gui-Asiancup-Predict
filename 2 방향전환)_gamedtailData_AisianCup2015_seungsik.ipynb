```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
b=pd.read_csv('./gamedetail_2015.csv', encoding = 'utf-8')
b.head()
```

```python
# 각 팀 데이터로 나누기
team1 = b[['gameday', 'team1', 'weather_d', 'shoot1', 'foul1', 'possession1']]
team2 = b[['gameday', 'team2', 'weather_d', 'shoot2', 'foul2', 'possession2']]
```

```python
# 팀 칼럼 동일하게 맞추기
team1.columns = ['gameday', 'team', 'weather_d', 'shoot', 'foul', 'possession']
team2.columns = ['gameday', 'team', 'weather_d', 'shoot', 'foul', 'possession']
```

```python
# 데이터 합치기
AsianGame2015 = pd.concat([team1,team2])
AsianGame2015.head()
```

```python
# 데이터 확인
AsianGame2015[AsianGame2015['gameday'] == "['31-Jan-15', '20:00']"]
```

```python
# 팀이름 변경??
len(AsianGame2015['team'].unique()) # 16개국
AsianGame2015['team'].unique()
```

```python
# 팀이름 한글로 변경
AsianGame2015.loc[AsianGame2015['team'] == "South Korea","team"] = '한국'
AsianGame2015.loc[AsianGame2015['team'] == "Iraq","team"] = '이라크'
AsianGame2015.loc[AsianGame2015['team'] == "Australia","team"] = '오스트레일리아'
AsianGame2015.loc[AsianGame2015['team'] == "Qatar","team"] = '카타르'
AsianGame2015.loc[AsianGame2015['team'] == "Kuwait","team"] = '쿠웨이트'
AsianGame2015.loc[AsianGame2015['team'] == "Japan","team"] = '일본'
AsianGame2015.loc[AsianGame2015['team'] == "Saudi Arabia","team"] = '사우디아라비아'
AsianGame2015.loc[AsianGame2015['team'] == "Oman","team"] = '오만'
AsianGame2015.loc[AsianGame2015['team'] == "North Korea","team"] = '북한'
AsianGame2015.loc[AsianGame2015['team'] == "Uzbekistan","team"] = '우즈베키스탄'
AsianGame2015.loc[AsianGame2015['team'] == "China, PR","team"] = '중국'
AsianGame2015.loc[AsianGame2015['team'] == "Jordan","team"] = '요르단'
AsianGame2015.loc[AsianGame2015['team'] == "Palestine","team"] = '팔레스타인'
AsianGame2015.loc[AsianGame2015['team'] == "Bahrain","team"] = '바레인'
AsianGame2015.loc[AsianGame2015['team'] == "Iran","team"] = '이란'
AsianGame2015.loc[AsianGame2015['team'] == "United Arab Emirates","team"] = '아랍에미리트'
AsianGame2015.head(10)
```

```python
# 날씨 기온 칼럼 생성

tempd = []
weathd = []
for i in AsianGame2015['weather_d']:
    temp = i.split()
    tempd.append(temp[-1])
    weathd.append(temp[-2])
```

```python
# 날씨 기온 칼럼 생성
AsianGame2015['temp'] = tempd
AsianGame2015['weather'] = weathd
AsianGame2015.head()
```

```python
# weater data 정제
temp2 = []
for i in AsianGame2015['temp']:
    temp2.append(i[0:2])
temp2
```

```python
AsianGame2015['temp2'] = temp2
AsianGame2015.head()
```

```python
# 날씨 데이터 살펴보기
AsianGame2015['weather'].unique()
```

```python
# 11년도 기준 날씨와 일치 시키기
AsianGame2015.loc[AsianGame2015['weather'] == 'Shower','weather'] = 'Rain'
AsianGame2015.loc[AsianGame2015['weather'] == 'Cloudy/Windy','weather'] = 'Cloudy'
AsianGame2015.loc[AsianGame2015['weather'] == 'Vicinity','weather'] = 'Rain'
AsianGame2015.loc[AsianGame2015['weather'] == 'Unknown','weather'] = 'Cloudy'
AsianGame2015.loc[AsianGame2015['weather'] == 'Precipitation','weather'] = 'Cloudy'
AsianGame2015.loc[AsianGame2015['weather'] == 'Drizzle','weather'] = 'Rain'
AsianGame2015.loc[AsianGame2015['weather'] == 'Fair','weather'] = 'Sunny'
```

```python
# 날씨 데이터 살펴보기
AsianGame2015['weather'].unique()
```
