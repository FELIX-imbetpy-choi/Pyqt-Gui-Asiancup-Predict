```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
b=pd.read_csv('./data/AsianCupgamedetail_2011_After.csv', encoding = 'utf-8')
b.tail()
```

```python
# 각 팀 데이터로 나누기
team1 = b[['gameday', 'team1', 'weather_d', 'shoot1', 'foul1', 'possession1']]
team2 = b[['gameday', 'team2', 'weather_d', 'shoot2', 'foul2', 'possession2']]
```

```python
# 팀 칼럼 동일하게 맞추기
team1.columns = ['gameday', 'team', 'weather_d', 'shoot', 'foul', 'possession']
team2.columns = ['gameday', 'team', 'weather_d', 'shoot', 'foul', 'possession']
```

```python
# 데이터 합치기
AsianGame2011 = pd.concat([team1,team2])
AsianGame2011.head()
```

```python
# 데이터 확인
AsianGame2011[AsianGame2011['gameday'] == "['07-Jan-11', '19:15']"]
```

```python
# 팀이름 변경??
len(AsianGame2011['team'].unique()) # 16개국
AsianGame2011['team'].unique()
```

```python
# 팀이름 한글로 변경
AsianGame2011.loc[AsianGame2011['team'] == "Qatar","team"] = '카타르'
AsianGame2011.loc[AsianGame2011['team'] == "Kuwait","team"] = '쿠웨이트'
AsianGame2011.loc[AsianGame2011['team'] == "Japan","team"] = '일본'
AsianGame2011.loc[AsianGame2011['team'] == "Saudi Arabia","team"] = '사우디아라비아'
AsianGame2011.loc[AsianGame2011['team'] == "India","team"] = '인도'
AsianGame2011.loc[AsianGame2011['team'] == "South Korea","team"] = '한국'
AsianGame2011.loc[AsianGame2011['team'] == "North Korea","team"] = '북한'
AsianGame2011.loc[AsianGame2011['team'] == "Iraq","team"] = '이라크'
AsianGame2011.loc[AsianGame2011['team'] == "Uzbekistan","team"] = '우즈베키스탄'
AsianGame2011.loc[AsianGame2011['team'] == "China, PR","team"] = '중국'
AsianGame2011.loc[AsianGame2011['team'] == "Jordan","team"] = '요르단'
AsianGame2011.loc[AsianGame2011['team'] == "Syria","team"] = '시리아'
AsianGame2011.loc[AsianGame2011['team'] == "Australia","team"] = '오스트레일리아'
AsianGame2011.loc[AsianGame2011['team'] == "Bahrain","team"] = '바레인'
AsianGame2011.loc[AsianGame2011['team'] == "Iran","team"] = '이란'
AsianGame2011.loc[AsianGame2011['team'] == "United Arab Emirates","team"] = '아랍에미리트'
AsianGame2011.head(10)
```

```python
# 날씨 기온 칼럼 생성

tempd = []
weathd = []
for i in AsianGame2011['weather_d']:
    temp = i.split()
    tempd.append(temp[-1])
    weathd.append(temp[-2])
```

```python
# 날씨 기온 칼럼 생성
AsianGame2011['temp'] = tempd
AsianGame2011['weather'] = weathd
AsianGame2011.head()
```

```python
# weater data 정제
temp2 = []
for i in AsianGame2011['temp']:
    temp2.append(i[0:2])
temp2
```

```python
AsianGame2011['temp2'] = temp2
AsianGame2011.head()
```

```python
# 날씨 데이터 살펴보기
AsianGame2011['weather'].unique()
```

```python
# 2015 년도 데이터와 일치시키기
AsianGame2011.loc[AsianGame2011['weather']=='Fair',"weather"] = "Sunny"
AsianGame2011.loc[AsianGame2011['weather']=='Haze',"weather"] = "Rain"
```
