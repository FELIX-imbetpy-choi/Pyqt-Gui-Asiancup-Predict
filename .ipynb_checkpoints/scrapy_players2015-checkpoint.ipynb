```python
from selenium import webdriver
import urllib.request as req
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome('./chromedriver')
driver.get('https://en.wikipedia.org/wiki/2015_AFC_Asian_Cup_squads')
```

```python
# 네임 추출 테스트
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
name = soup.select('#mw-content-text > div > table > tbody > tr > th > a')[22].text
age = soup.select('#mw-content-text > div > table > tbody > tr > td')[2].text.split()[4].replace(')','')
cap = soup.select('#mw-content-text > div > table > tbody > tr > td')[3].text.split()[0]
goal = soup.select('#mw-content-text > div > table > tbody > tr > td')[4].text
club = soup.select('#mw-content-text > div > table > tbody > tr > td')[5].text
print(name)
print(age)
print(cap)
print(goal)
print(club)
```

```python
# 부상자 때문에 엔트리가 한명 적어진 인도는 따로 Scarpy
nameList = []
ageList = []
capList = []
goalList = []
clubList = []

html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
for i in range(0,367):
    name = soup.select('#mw-content-text > div > table > tbody > tr > th > a')[i].text
    nameList.append(name)
for i in range(0,2202):
    if i % 6 == 2:
        age = soup.select('#mw-content-text > div > table > tbody > tr > td')[i].text.split()[4].replace(')','')
        ageList.append(age)    
    if i % 6 == 3:
        cap = soup.select('#mw-content-text > div > table > tbody > tr > td')[i].text.strip()
        capList.append(cap)
    if i % 6 == 4:
        goal = soup.select('#mw-content-text > div > table > tbody > tr > td')[i].text.strip()
        goalList.append(goal)
    if i % 6 == 5:
        club = soup.select('#mw-content-text > div > table > tbody > tr > td')[i].text.strip()
        clubList.append(club)
```

```python
print(nameList)
print(ageList)
print(capList)
print(goalList)
print(clubList)
```

```python
players_2015 = pd.DataFrame(ageList)
players_2015 = players_2015.rename(columns = {0:'age'})
players_2015['cap'] = capList
players_2015['goal'] = goalList
```

```python
# goal 결측치 제거
players_2015[players_2015['goal']=='']
```

```python
players_2015.loc[players_2015['goal']=='','goal'] = 0
```

```python
players_2015 = players_2015.astype(int)
```

```python
players_2015['name'] = nameList
```

```python
players_2015['club'] = clubList
```
