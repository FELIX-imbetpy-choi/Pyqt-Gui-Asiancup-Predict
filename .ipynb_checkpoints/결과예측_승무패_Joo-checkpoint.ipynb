```python
import pymysql
import pandas as pd
```

```python
# mysql Connection
conn = pymysql.Connect(host = '192.168.101.106', user = 'root', password = '1111',
                       db = 'asiancup', charset = 'utf8')
```

```python
# Connection으로부터 cursor 생성
curs = conn.cursor()
```

```python
# SQL 문장
sql = 'select * from nations2011'
curs.execute(sql)
```

```python
# Data Fetch
rows = curs.fetchall()
print(rows)
```

```python
rowList = list(rows)
```

```python
print(rowList)
```

```python
total2011 = pd.DataFrame(rowList)
```

```python
# 칼럼 이름 변경
total2011.columns = ['seq','gameday','team','shoot','foul','possession','weater','temp','rank','salary']
total2011.head()
```

```python
# 날짜 순으로 내림차순
total2011 = total2011.sort_values(by='gameday', ascending=False)
```

```python
# SQL 문장
sql = 'select * from nations2015'
curs.execute(sql)
```

```python
# Data Fetch
rows = curs.fetchall()
print(rows)
```

```python
rowList = list(rows)
```

```python
total2015 = pd.DataFrame(rowList)
```

```python
# 칼럼 이름 변경
total2015.columns = ['seq','gameday','team','shoot','foul','possession','weater','temp','rank','salary']
total2015.head()
```

```python
total2015 = total2015.sort_values(by='gameday', ascending=False)
```
