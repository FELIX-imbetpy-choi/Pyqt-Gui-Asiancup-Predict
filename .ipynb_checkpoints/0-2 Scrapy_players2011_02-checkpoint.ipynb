```python
from selenium import webdriver
import urllib.request as req
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome('./chromedriver')
driver.get('https://www.transfermarkt.co.uk/usbekistan/startseite/verein/3563/saison_id/2010')
```

```python
# 네임 추출 테스트
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
backnum = soup.select('#yw1 > table > tbody > tr > td ')[0].text.strip()
name = soup.select('#yw1 > table > tbody > tr > td ')[1].text.strip()
age = soup.select('#yw1 > table > tbody > tr > td')[29].text.split()[0].replace('(','').replace(')','')
height = soup.select('#yw1 > table > tbody > tr > td ')[4].text.strip().replace(',','').replace('m','')
cap = soup.select('#yw1 > table > tbody > tr > td ')[6].text.strip()
goal = soup.select('#yw1 > table > tbody > tr > td ')[7].text.strip()
salary = soup.select('#yw1 > table > tbody > tr > td ')[9].text.strip().replace('£','').replace('k','000')
print(backnum)
print(name)
print(age)
print(height)
print(cap)
print(goal)
print(salary)
```

```python
# 부상자 때문에 엔트리가 두명 적어진 한국은 따로 Scarpy
backnumList = []
nameList = []
ageList = []
heightList = []
capList = []
goalList = []
salaryList = []

html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
for i in range(0,330):
    if i % 10 == 1:
        name = soup.select('#yw1 > table > tbody > tr > td')[i].text.split()
        nameList.append(name)
    if i % 10 == 2:
        age = soup.select('#yw1 > table > tbody > tr > td')[i].text.split()[3].replace('(','').replace(')','')
        ageList.append(age)    
    if i % 10 == 4:
        height = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace(',','').replace('m','').strip()
        heightList.append(height)
    if i % 10 == 6:
        cap = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('-','0').strip()
        capList.append(cap)
    if i % 10 == 7:
        goal = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('-','0').strip()
        goalList.append(goal)
    if i % 10 == 9:
        salary = soup.select('#yw1 > table > tbody > tr > td')[i].text.replace('.','').replace('£','').replace('k','').replace('m','000').strip()
        salaryList.append(salary)
```

```python
print(nameList)
print(ageList)
print(heightList)
print(capList)
print(goalList)
print(salaryList)
```

```python
players = pd.DataFrame(heightList)
players = players.rename(columns = {0:'height'})
players['age'] = ageList
players['cap'] = capList
players['goal'] = goalList
players['salary'] = salaryList
```

```python
# players['age'] = players['age'].str[-3:-1]
```

```python
players.loc[players['height']==''] = 0
players.loc[players['height']=='N/A','height'] = 0
players.loc[players['age'].isnull(),'age'] = 0
players.loc[players['salary']=='-\xa0'] = 0
players.loc[players['salary']=='-','salary'] = 0
```

```python
players
```

```python
players = players.astype(int)
```

```python
# 키 정보가 없는 선수들의 키를 평균값으로 대체
players.loc[players['height']==0,'height'] =int(players['height'][players['height']!=0].mean())
```

```python
# salary 정보가 없는 선수들의 연봉을 평균값으로 대체
players.loc[players['salary']==0,'salary'] =int(players['salary'][players['salary']!=0].median())
```

```python
# 나이 정보가 없는 선수들의 키를 평균값으로 대체
players.loc[players['age']==0,'age'] =int(players['age'][players['age']!=0].mean())
```
