```python
aList = soup.select('#store_list > table > tbody > tr > td:nth-of-type(2)') # 매장이름
bList = soup.select('#store_list > table > tbody > tr > td:nth-of-type(3)') # 매장 전번
cList = soup.select('#store_list > table > tbody > tr > td:nth-of-type(4) > p > a') # 매장 코드

store = []
tel = []
storecord=[] # store code - 이용하여 상세보기 페이지 주소 가져오려고 (매장 지번이 매장코드를 통해 검색가능함)
list1 = []

for k in aList:
    aa=k.string
    store.append(aa.strip())

for i in bList:
    tel.append(i.string)
    
for a in cList:
    bb=a['onclick'].replace("goStore('","")
    cc=bb.replace("')","")
    storecord.append(cc)
    
no = 0
for z in store:
    list1.append([z,tel[no],storecord[no]])
    no += 1
```

# 받아온 데이터 정제, 데이터프레임 형성 -> 매장코드로 주소 데이터 불러오기

```python
store3 = []
count1 = 0
for a in range(0,103):
    for b in range(0,10):
        count1+=1
        conv1=clickstore[a][b].get_text()
        store3.append(conv1.strip())
```

```python
storetel3 = []
count1 = 0
for a in range(0,103):
    for b in range(0,10):
        count1+=1
        conv1=clicktel[a][b].get_text()
        storetel3.append(conv1.strip())
storetel3
```

```python
storecode3 = []
count1 = 0
for a in range(0,103):
    for b in range(0,10):
        count1+=1
        conv1=clickcode[a][b]
        conv2=conv1['onclick'].replace("goStore('","")
        conv3=conv2.replace("')","")
        storecode3.append(conv3.strip())
storecode3
```

```python
# 매장 코드 이용하여 주소 가져오기
import urllib.request as req

# 매장 주소 리스트
staddresslist=[]

for scode in storecode3:
    url = 'http://www.twosome.co.kr/store/storeView.asp?sort_store=%s'%scode
    res = req.urlopen(url)
    soup = BeautifulSoup(res, 'html.parser')
    time.sleep(2)
    staddress = soup.select('#content > table > tbody > tr:nth-of-type(2) > td')
    staddress2 = staddress
    staddresslist.append(staddress2)

print(staddresslist)
```

```python
twoaddress=pd.read_csv('./data/twosome_address_1029.csv',index_col=0)
```

```python
# 투썸의
two_lng = [] # 투썸의 위도
two_lat = [] # 투썸의 경도

# 사용목적에 맞게 수정
import os
import sys
import urllib.request
import json
client_id = "FhouMGJbmgH2XI8LgCYa"
client_secret = "jqJ0ZpqL8f"

for name in twoaddress['주소']:
    print(name)
    encText = urllib.parse.quote(name)
    url = "https://openapi.naver.com/v1/map/geocode?query=" + encText # json 결과
    # url = "https://openapi.naver.com/v1/map/geocode.xml?query=" + encText # xml 결과
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
        #print(response_body.decode('utf-8'))
        entext = response_body.decode('utf-8')
    else:
        print("Error Code:" + rescode)
    #---------------------------
    data = json.loads(entext)
    lng = data['result']['items'][0]['point']['x']
    lat = data['result']['items'][0]['point']['y']
    two_lng.append(lng)
    two_lat.append(lat)
```

```python
lnglat=pd.DataFrame({
    'lng' : two_lng,
    'lat' : two_lat,
    'brand' : 'TWOSOME PLACE',
    '주소' : twoaddress['주소']
})
```

```python
lnglat.to_csv('./data/TWOSOME_LNG_LAT')
```
