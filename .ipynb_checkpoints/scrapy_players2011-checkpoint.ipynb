```python
from selenium import webdriver
import urllib.request as req
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver = webdriver.Chrome('./chromedriver')
driver.get('https://en.wikipedia.org/wiki/2011_AFC_Asian_Cup_squads')
```

```python
# 네임 추출 테스트
html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
name = soup.select('#mw-content-text > div > table > tbody > tr > th > a')[346]
age = soup.select('#mw-content-text > div > table > tbody > tr > td')[2].text.split()[4].replace(')','')
cap = soup.select('#mw-content-text > div > table > tbody > tr > td')[3].text.strip()
goal = soup.select('#mw-content-text > div > table > tbody > tr > td')[4].text.strip()
print(name)
print(age)
print(cap)
print(goal)
print(club)
```

```python
# 부상자 때문에 엔트리가 한명 적어진 인도는 따로 Scarpy
nameList = []
ageList = []
capList = []
goalList = []
clubList = []

html = driver.page_source
soup = BeautifulSoup(html,'html.parser')
for i in range(0,368):
    name = soup.select('#mw-content-text > div > table > tbody > tr > th > a')[i].text
    nameList.append(name)
for i in range(0,1840):
    if i % 5 == 2:
        age = soup.select('#mw-content-text > div > table > tbody > tr > td')[i].text.split()[4].replace(')','')
        ageList.append(age)    
    if i % 5 == 3:
        cap = soup.select('#mw-content-text > div > table > tbody > tr > td')[i].text.strip()
        capList.append(cap)
    if i % 5 == 4:
        club = soup.select('#mw-content-text > div > table > tbody > tr > td')[i].text.strip()
        clubList.append(club)
```

```python
print(nameList)
print(ageList)
print(capList)
print(clubList)
```

```python
players_2011 = pd.DataFrame(ageList)
players_2011 = players_2011.rename(columns = {0:'age'})
players_2011['cap'] = capList
```
